```python
#Author: Karen Amadi
#Assignment1
# Variable declarations with data type
gym_member = "Alex Alliton"  # str
preferred_weight_kg = 20.5  # float
highest_reps = 25  # int
membership_active = True  # bool

# Dictionary to store workout statistics of friends
# Data type: dict[str, tuple[int, int, int]]
workout_stats = {
    "Alex": (30, 45, 20),
    "Jamie": (40, 50, 30),
    "Taylor": (35, 40, 25)
}

# Calculate total workout minutes and store separately
workout_totals = {friend: sum(minutes) for friend, minutes in workout_stats.items()}

# Update the original dictionary with total workout minutes
for friend, total_minutes in workout_totals.items():
    workout_stats[f"{friend}_Total"] = total_minutes

# 2-dimensional list (nested list) of workout minutes
# Data type: list[list[int]]
workout_list = [list(minutes) for friend, minutes in workout_stats.items() if not friend.endswith("_Total")]

# Slicing the workout_list
# Extracting yoga and running minutes for all friends
yoga_running_minutes = [row[:2] for row in workout_list]
print("Yoga and Running Minutes for all friends:", yoga_running_minutes)

# Extracting weightlifting minutes for the last two friends
weightlifting_last_two = [row[2] for row in workout_list[-2:]]
print("Weightlifting Minutes for last two friends:", weightlifting_last_two)

# Checking if any friend's total workout minutes are >= 120
for friend, total in workout_stats.items():
    if friend.endswith("_Total") and total >= 120:
        name = friend.replace("_Total", "")
        print(f"Great job staying active, {name}!")

# Test values for friend_name instead of input
friend_name = input("Enter friend's name ------ ")
if friend_name in workout_stats:
    activities = workout_stats[friend_name]
    total = workout_stats.get(f"{friend_name}_Total", 0)
    print(f"{friend_name}'s workout minutes:")
    print(f"Yoga: {activities[0]}, Running: {activities[1]}, Weightlifting: {activities[2]}")
    print(f"Total Minutes: {total}")
else:
    print(f"Friend {friend_name} not found in the records.")

# Finding the friend with highest and lowest total workout minutes
friends_totals = {friend.replace("_Total", ""): total for friend, total in workout_stats.items() if friend.endswith("_Total")}
highest_friend = max(friends_totals, key=friends_totals.get)
lowest_friend = min(friends_totals, key=friends_totals.get)

print(f"Friend with highest total workout minutes: {highest_friend} ({friends_totals[highest_friend]} minutes)")
print(f"Friend with lowest total workout minutes: {lowest_friend} ({friends_totals[lowest_friend]} minutes)")
```
